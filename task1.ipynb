# Задание

Построить графики абсолютной погрешности каждого метода в зависимости от шага численного дифференцирования $$ h_n = \frac{n}{2^n},  n \in [1, 21] $$

## Функции
$$ sin (x^2) \\
cos (sin (x)) \\ 
exp (sin (cos (x))) \\
ln (x + 3) \\
\sqrt{x + 3}
$$

Использовать логарифмический масштаб по обеим осям. 

Графики погрешностей методов для каждой функции строить в отдельном окне.

## Список формул
$$

\frac {f(x+h) - f(x)}{h} \\ 

\frac {f(x) - f(x-h)} {h} \\

\frac {f(x+h) - f(x-h)} {2h} \\

\frac{4}{3} \frac {f(x+h) - f(x-h)} {2h} - \frac{1}{3} \frac {f(x+2h) - f(x-2h)}{4h} \\

\frac{3}{2} \frac{f(x+h) - f(x-h)}{2h} - \frac{3}{5} \frac{f(x+2h) - f(x-2h)}{2h} + \frac{1}{10} \frac {f(x+3h) - f(x-3h)}{6h}

$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

h_number = 21

list_h = np.zeros(0)
for n in range (1, h_number + 1):
    list_h = np.append (list_h, n / 2**n)

print (list_h)

x = np.pi / 3
```

```python
def method_1 (func, exact_value):
    return abs (exact_value - (func (x + list_h) - func(x)) / list_h)

def method_2 (func, exact_value):
    return abs (exact_value - (func(x) - func(x - list_h)) / list_h)

def method_3 (func, exact_value):
    return abs (exact_value - (func(x + list_h) - func(x - list_h)) / (2 * list_h))

def method_4 (func, exact_value):
    return abs (exact_value - (4/3 * (func (x +     list_h) - func (x -     list_h)) / (2 * list_h) - 
                               1/3 * (func (x + 2 * list_h) - func (x - 2 * list_h)) / (4 * list_h)))

def method_5 (func, exact_value):
    return abs (exact_value - (3/2 * (func (x +     list_h) - func (x -     list_h)) / (2 * list_h) - 
                               3/5 * (func (x + 2 * list_h) - func (x - 2 * list_h)) / (4 * list_h) + 
                               1/10* (func (x + 3 * list_h) - func (x - 3 * list_h)) / (6 * list_h)))
```

### $$ sin (x ^ 2) $$

```python
def func_1 (x):
    return np.sin(x ** 2)

exact_value = 0.9563079109495478438213288631970314825236940468782193760660893599

first  = method_1 (func_1, exact_value)
second = method_2 (func_1, exact_value)
third  = method_3 (func_1, exact_value)
fourth = method_4 (func_1, exact_value)
fifth  = method_5 (func_1, exact_value)

plot_info = np.array ([first, second, third, fourth, fifth])
plt.figure(figsize=(15,9))
plt.xlabel ("ln(h)")
plt.ylabel ("ln (error)")
for method_values in plot_info:
    plt.loglog (list_h, method_values, marker="P")
```
